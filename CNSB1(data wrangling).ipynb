```python
import datetime as DT
import io
import numpy as np
import pandas as pd
now = pd.Timestamp('now')
data1['DateOpen'] = pd.to_datetime(data1['DateOpen'], format='%m%d%y')    # 1
data1['DateOpen'] = data1['DateOpen'].where(data1['DateOpen'] < now, data1['DateOpen'] -  np.timedelta64(100, 'Y'))   # 2
data1['DateOpen'] = (now - data1['DateOpen']).astype('<m8[Y]')    # 3
pd.options.mode.chained_assignment = 'warn'
#print(df['age'])
```
